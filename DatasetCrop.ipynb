```python
import cv2
import matplotlib.pyplot as plt
```

```python
folder = ""
bacteria = "positive"
image = "028"
category = ["png", 'jpg', 'jpeg']
imtype = category[1]
# top, bottom, right, left
margin = [100, 0, 0, 0]
img = cv2.imread(folder+bacteria+"/"+bacteria+image+"."+imtype)
img = img[margin[0]:(img.shape[0]-margin[1]), margin[3]:(img.shape[1]-margin[2])]

# just to show the image
toshow = plt.imread(bacteria+"/"+bacteria+image+"."+imtype)
plt.imshow(toshow[margin[0]:(img.shape[0]-margin[1]), margin[3]:(img.shape[1]-margin[2])])
```

```python
#### start = 0
height = 100
width = 100
cut = 50
x = int((img.shape[1] - width)/cut + 1)
y = int((img.shape[0] - height)/cut + 1)
print("you will get", x*y, "images")
```

```python
counter = 0
folder = "data/"
for hcut in range(x):
    for vcut in range(y):
        counter+=1
        imcut = img[ int(vcut*cut):int(vcut*cut+height), int(hcut*cut):int(hcut*cut + width)]
        cv2.imwrite(folder+bacteria+"/"+bacteria+image+str(vcut)+str(hcut)+".jpg", imcut)
#         print( vcut*cut, ":", vcut*cut+100,",", hcut*cut,":",hcut*cut + 100)
print("created", counter, "images of dimension",width,"x",height)
```
